## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("Resources/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
a_df = application_df.drop(['EIN', 'NAME'], axis=1)
a_df.head()
```

```python
# Determine the number of unique values in each column.
counts = a_df.nunique()
counts
```

```python
# Look at APPLICATION_TYPE value counts for binning
counts = a_df['APPLICATION_TYPE'].value_counts()
counts
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
cut = 45
app_count = dict(counts)
application_types_to_replace = []

# iterate through items in dictionary, add app types with value < cutoff to list
for x, value in app_count.items():
    if value < cut:
        application_types_to_replace.append(x)

# Replace in dataframe
for app in application_types_to_replace:
    a_df['APPLICATION_TYPE'] = a_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
a_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
class_counts = a_df['CLASSIFICATION'].value_counts()
class_counts
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
class_counts.loc[class_counts > 1]
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
cut = 75
class_count = dict(class_counts)
classifications_to_replace = []

# iterate through items in dictionary, add classifications with value < cutoff to list
for x, value in class_count.items():
    if value < cut:
        classifications_to_replace.append(x)


# Replace in dataframe
for cls in classifications_to_replace:
    a_df['CLASSIFICATION'] = a_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
a_df['CLASSIFICATION'].value_counts()
```

```python
# Execute PD Dummies 
dum_df = pd.get_dummies(a_df)
dum_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
X = dum_df.drop('IS_SUCCESSFUL', axis=1).values
y = dum_df['IS_SUCCESSFUL'].values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=80)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
#Get number of inputs
number_input = len(X_train[0])
number_input
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=60, activation="relu", input_dim=number_input))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=40, activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs=75)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
